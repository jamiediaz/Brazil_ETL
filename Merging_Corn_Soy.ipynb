```python
#importing libraries
import os
import glob
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
pd.set_option('display.max_columns', 500)
```

```python
#loading csv file
#brazilian government expenses, including all the ones related to agriculture or land
#obtained from https://www.kaggle.com/vicentegsantos/brazil-gov-expenses
#the original file (Brazil_Gov_Exp_All - Formated.csv) was not in CSV format, so it was converted to Brazil.csv in Excel
gov_file = "./Resources/Brazil.csv"
```

```python
#reading csv
gov_df = pd.read_csv(gov_file, encoding='ISO-8859-1')
gov_df.head()
```

```python
#soy bean production/consupmtion totals per year, including global and brazilian numbers
#saved all the soy files to "soy" subfolder, and all the corn files to "corn" subfolder
#obtained from https://www.kaggle.com/ainslie/usda-wasde-monthly-corn-soybean-projections

#switching to soy directory (put all the soy files in "soy folder")
os.chdir("./Resources/soy_corn/soy")

#loading csv files
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
combined_soy_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_soy_csv.to_csv( "combined_soy_csv.csv", index=False, encoding='utf-8-sig')
```

```python
combined_soy_csv.head()
```

```python
#Saving soy total values on global dataframe
combined_soy_csv_df = pd.DataFrame(combined_soy_csv)
combined_soy_global_df = combined_soy_csv_df[["Date", "Area Planted", "Yield", "Area Harvested", "Beginning Stocks", \
                                           "Production", "Imports", "Total Supply", "Crushings", "Exports"]]
combined_soy_global_df.head()
```

```python
#Saving soy Brazilian values on Brazil dataframe
combined_soy_brazil_df = combined_soy_csv_df[["Date", "Brazil BeginningStocks","Brazil Production","Brazil Imports", \
                                              "Brazil DomesticTotal", "Brazil DomesticCrush","Brazil Exports", "Brazil EndingStocks"]]
combined_soy_brazil_df.head()
```

```python
#switching to corn directory (put all the soy files in "corn folder")
os.chdir("../corn")
```

```python
#loading csv files
#corn production/consupmtion totals per year, including global and brazilian numbers
#obtained from https://www.kaggle.com/ainslie/usda-wasde-monthly-corn-soybean-projections
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

#combine all files in the list
combined_corn_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_corn_csv.to_csv( "combined_corn_csv.csv", index=False, encoding='utf-8-sig')
```

```python
combined_corn_csv.head()
```

```python
#Saving corn total values on global dataframe
combined_corn_csv_df = pd.DataFrame(combined_corn_csv)
combined_corn_global_df = combined_corn_csv_df[["Date", "Area Planted", "Yield", "Area Harvested", "Beginning Stocks", \
                                           "Production", "Imports", "Total Supply", "Feed and Residual", "Exports"]]
combined_corn_global_df.head()
```

```python
#Saving corn Brazilian values on Brazil dataframe
combined_corn_brazil_df = combined_corn_csv_df[["Date", "Brazil BeginningStocks","Brazil Production","Brazil Imports", \
                                              "Brazil DomesticTotal", "Brazil DomesticFeed","Brazil Exports", "Brazil EndingStocks"]]
combined_corn_brazil_df.head()
```

```python
#uploading
engine = create_engine('postgresql://sqladmin:password@localhost:5432/brazil_etl')
combined_soy_brazil_df.to_sql('combined_soy_brazil', engine, if_exists='replace', index=False)
```

```python
#uploading do postgres sql db
engine = create_engine('postgresql://sqladmin:password@localhost:5432/brazil_etl')
combined_soy_brazil_df.to_sql('combined_soy_brazil', engine, if_exists='replace', index=False)
```

```python
engine = create_engine('postgresql://sqladmin:password@localhost:5432/brazil_etl')
gov_df.to_sql('gov_expenses_brazil', engine, if_exists='replace', index=False)
```

```python
engine = create_engine('postgresql://sqladmin:password@localhost:5432/brazil_etl')
combined_corn_brazil_df.to_sql('combined_corn_brazil', engine, if_exists='replace', index=False)
```
